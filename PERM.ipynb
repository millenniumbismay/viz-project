```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
perm_data = pd.read_excel('./data/perm_data/PERM_FY2012.xlsx')
perm_data.head()
```

```python
print(list(perm_data.columns))
```

```python
perm_data['US_ECONOMIC_SECTOR'].unique()
```

```python
# columns_to_keep = ['CASE_STATUS', 'EMPLOYER_NAME', 'EMPLOYER_STATE',
#                    'US_ECONOMIC_SECTOR', 'PW_SOC_TITLE', 'PW_AMOUNT_9089',
#                    'PW_UNIT_OF_PAY_9089', 'COUNTRY_OF_CITZENSHIP',
#                    'CLASS_OF_ADMISSION'] 2010 - 2012
columns_to_keep = ['CASE_STATUS', 'EMPLOYER_NAME', 'EMPLOYER_STATE',
                   'US_ECONOMIC_SECTOR', 'PW_SOC_TITLE', 'PW_AMOUNT_9089',
                   'PW_UNIT_OF_PAY_9089', 'COUNTRY_OF_CITIZENSHIP',
                   'CLASS_OF_ADMISSION']
perm_data_filtered = perm_data[columns_to_keep]
perm_data_filtered.head()
```

```python
perm_data_filtered.shape
```

```python
perm_data_cleaned = perm_data_filtered.dropna(subset=['Class_of_Admission'])
perm_data_cleaned.shape
```

```python
perm_data_cleaned['Class_of_Admission'].unique()
```

```python
perm_final = perm_data_cleaned[perm_data_cleaned['Class_of_Admission'].str.startswith('H', na=False)]
perm_final.head()
```

```python
perm_final['Class_of_Admission'].unique()
```

```python
perm_final.shape
```

```python
perm_final.to_csv('./data/perm_data/perm_2014.csv')
```
