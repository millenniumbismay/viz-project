```python
niv_excel = pd.ExcelFile('./data/FYs97-22_NIVDetailTable.xlsx')
```

```python
worksheets = ['FY10', 'FY11', 'FY12', 'FY13', 'FY14', 'FY15', 'FY16', 'FY17',
               'FY18', 'FY19', 'FY20', 'FY21', 'FY22']
for worksheet in worksheets:
  df = pd.read_excel(niv_excel, worksheet)
  # print(df.columns, df.columns.nunique())
  # print(df.shape)
  columns_to_keep = []
  for column in list(df.columns):
    if column.startswith('O') or column.startswith('E') or column.startswith('H') or column.startswith('Fisc'):
      columns_to_keep.append(column)
  print(columns_to_keep)
  df = df[columns_to_keep]
  print(df.shape)
  filtered_data_path = "./data/niv_data/niv_"+worksheet+".csv"
  df.to_csv(filtered_data_path)
  print("*"*50)
```
